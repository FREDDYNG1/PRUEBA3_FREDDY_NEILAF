```python
personas = []

def GUARDAR():
    dni = input("Ingrese su DNI, sin puntos ni guion: ")
    nombre = input("Ingrese NOMBRE: ")
    edad = int(input("Ingrese EDAD: "))
    estado_civil = input("Ingrese su ESTADO CIVIL: ")
    pais_de_nacimiento = input("Ingrese PAÍS DE NACIMIENTO: ")
    ciudad_de_nacimiento = input("Ingrese CIUDAD DE NACIMIENTO: ")

    if validar_dni(dni):
        persona = {
            "dni": dni,
            "nombre": nombre,
            "edad": edad,
            "estado_civil": estado_civil,
            "pais_de_nacimiento": pais_de_nacimiento,
            "ciudad_de_nacimiento": ciudad_de_nacimiento
        }
        personas.append(persona)
        print("Persona guardada exitosamente.")
    else:
        print("El DNI ingresado no es válido.")

def validar_dni(dni):
    dni = dni.replace(".", "").replace("-", "")
    if len(dni) != 9 or not dni[:-1].isdigit():
        return False
    multiplicadores = [2, 3, 4, 5, 6, 7, 2, 3]
    total = sum(int(digito) * multiplicador for digito, multiplicador in zip(dni[:-1][::-1], multiplicadores))
    resto = total % 11
    digito_verificador = str(11 - resto)
    if digito_verificador == "11":
        digito_verificador = "0"
    elif digito_verificador == "10":
        digito_verificador = "K"
    return dni[-1].upper() == digito_verificador

def buscar():
    dni = input("Ingrese el DNI de la persona que desea buscar: ")
    found = False
    for persona in personas:
        if persona["dni"] == dni:
            print("Persona encontrada:")
            if persona["pais_de_nacimiento"].lower() == "argentina":
                print("La persona pertenece a la República Argentina.")
            else:
                print("La persona NO pertenece a la República Argentina.")
            print("DNI:", persona["dni"])
            print("Nombre:", persona["nombre"])
            print("Edad:", persona["edad"])
            print("Estado civil:", persona["estado_civil"])
            print("País de Nacimiento:", persona["pais_de_nacimiento"])
            print("Ciudad de Nacimiento:", persona["ciudad_de_nacimiento"])
            found = True
            break
    if not found:
        print("Persona no encontrada.")

def imprimir_certificados():
    dni = input("Ingrese el DNI de la persona para la impresión del certificado: ")
    found = False
    for persona in personas:
        if persona["dni"] == dni:
            print("¿Qué certificado desea imprimir?")
            print("1. Certificado de nacimiento")
            print("2. Certificado de matrimonio")
            opcion = input("Ingrese el número de opción: ")

            if opcion == "1":
                imprimir_certificado_nacimiento(persona)
            elif opcion == "2":
                imprimir_certificado_matrimonio(persona)
            else:
                print("Opción inválida.")
            break


def imprimir_certificado_nacimiento(persona):
    print("Imprimiendo certificado de nacimiento de", persona["nombre"])
    print("Persona encontrada:")
    print("DNI:", persona["dni"])
    print("Nombre:", persona["nombre"])
    print("Edad:", persona["edad"])
    print("Estado civil:", persona["estado_civil"])
    print("País de Nacimiento:", persona["pais_de_nacimiento"])
    print("Ciudad de Nacimiento:", persona["ciudad_de_nacimiento"])


def imprimir_certificado_matrimonio(persona):
    print("Imprimiendo certificado de matrimonio de", persona["nombre"])
    print("Persona encontrada:")
    print("DNI:", persona["dni"])
    print("Nombre:", persona["nombre"])
    print("Edad:", persona["edad"])
    print("Estado civil:", persona["estado_civil"])
    print("País de Nacimiento:", persona["pais_de_nacimiento"])
    print("Ciudad de Nacimiento:", persona["ciudad_de_nacimiento"])


def eliminar():
    dni = input("Ingrese el DNI de la persona que desea eliminar: ")
    removed = False
    for persona in personas:
        if persona["dni"] == dni:
            personas.remove(persona)
            removed = True
            print("----------------------------------------")
            print("\tDATOS ELIMINADOS EXITOSAMENTE")
            print("----------------------------------------")
            break
        if not removed:
           print("Persona no encontrada.")

def main():
    while True:
        print("-----------------------------------------")
        print("\tBIENVENIDO A LOS SERVICIOS DE RENAPER")
        print("-----------------------------------------")
        print("\tQUÉ OPERACIÓN DESEA REALIZAR")
        print("-----------------------------------------")
        print("1. GUARDAR")
        print("2. BUSCAR")
        print("3. IMPRIMIR CERTIFICADOS")
        print("4. ELIMINAR")
        print("5. SALIR")

        opcion = int(input("Ingrese la opción que desea: "))

        if opcion == 1:
            GUARDAR()
        elif opcion == 2:
            buscar()
        elif opcion == 3:
            imprimir_certificados()
            print("----------------------------------------")
            print("\tCERTIFICADOS IMPRESOS CON ÉXITO")
            print("----------------------------------------")
        elif opcion == 4:
            eliminar()
        elif opcion == 5:
            print("----------------------------------------")
            print("\t\tHASTA LUEGO")
            print("GRACIAS POR PREFERIR LOS SERVICIOS DE RENAPER")
            print("----------------------------------------")
            print("\tFreddy NEilaf \t\tversion 1.0")
            print("----------------------------------------")
            break
        else:
            print("Opción inválida. Por favor, ingrese nuevamente.")

if __name__ == "__main__":
    main()

```
